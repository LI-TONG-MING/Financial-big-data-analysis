##產生文字串流
根據預設，模型會在完成整個文字產生程序後傳回回應。您可以不等待整個結果，改用串流處理部分結果，以便加快互動速度。

以下範例說明如何使用 streamGenerateContent 方法實作串流，從純文字輸入提示產生文字。

```python
response = model.generate_content("請分析近期的台積電股票與日月光投控股票的相關性", stream=True)
for chunk in response:
    print(chunk.text)
    print("_" * 80)
```

## 設定文字產生
您傳送至模型的每個提示都包含參數，用來控制模型產生回應的方式。您可以使用 GenerationConfig 設定這些參數。如果您未設定參數，模型會使用預設選項，這些選項可能因模型而異。

以下範例說明如何設定多個可用選項。

```python
response = model.generate_content(
    "請分析近期的台積電股票與日月光投控股票的相關性.",
    generation_config=genai.types.GenerationConfig(
        # Only one candidate for now.  ##目前只有一名候選人。
        candidate_count=1,  #candidateCount 指定要傳回的產生回應數量。目前，這個值只能設為 1。如果未設定，則預設值為 1。
        stop_sequences=["X"],  #stopSequences 指定將停止產生輸出的字元序列 (最多 5 個)。如果已指定，API 會在 stop_sequence 首次出現時停止。停止序列不會納入回應。
        max_output_tokens=100,   #maxOutputTokens 會設定候選項中可納入的符記數量上限。 ##  多思這個是字的意思吧!
        temperature=2.0,  #temperature 會控制輸出的隨機程度。如要取得更多創意回覆，請使用較高的值；如要取得較多確定性回覆，請使用較低的值。值的範圍為 [0.0, 2.0]。
    ),
)

print(response.text)
```

# 抓取線上股價資訊，由gemin API 判斷是否要買

```python
import yfinance as yf
taiwan_3711_stock = yf.Ticker("3711.TW")
data  = taiwan_3711_stock.history(period="2y")
```

```python
print(type(data))
print(data)
```

```python
data.to_csv('stock_csv_3711.csv', index=True)
```

```python
model = genai.GenerativeModel(model_name = "gemini-1.5-pro")
sample_csv = genai.upload_file("stock_csv_3711.csv")
response = model.generate_content(["請說明資料內容.", sample_csv])
print(response.text)
```

## 使用prompt 的寫法

```python

model = genai.GenerativeModel(model_name = "gemini-1.5-pro")
sample_csv = genai.upload_file("stock_csv_3711.csv")
prompt = "請分析這支股票好不好，我現在是不是該買，請以簡短的方式回覆"
response = model.generate_content([prompt, sample_csv])
print(response.text)
```

## 提供對股票的投資建議、評分排序與最佳投資標的選擇
